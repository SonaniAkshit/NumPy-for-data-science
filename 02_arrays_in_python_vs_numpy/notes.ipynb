*Operations on array*

```python
# iterate an array
a = arr.array('i',[10,20,30,40,50])

for x in a:
    print(x,end=" ")
```

```python
# indexed array elements
i = 0

while i < len(a):
    print(f"{i}:{a[i]}")
    i+=1
```

```python
# adding new element on array

#insert

a.insert(0,220)
print(*a)
```

```python
# remove selected element
a.remove(220)
print(*a)
```

```python
# append new element 
a.append(110)
print(*a)
```

```python
# delete from index
del a[5]
print(*a)
```

```python
# pop from last

a.pop()
print(*a)
```

```python
# pop from selecet index
a.pop(1)
print(*a)
```

```python
a = arr.array('i',[10,20,30,40,50])
print(*a)
```

```python
# Slicing
print(*a[::-1])
```

```python
print(*a[0:2:1])
```

```python
print(*a[2:4])
```

```python
# searching

print(a.index(40)) #it returns index of first occurrence specified element
```

```python
# edit

a[3] = 400
print(a[3])
print(*a)
```

```python
# counting 

a.count(10)
```

```python
a = arr.array('i',[1,2,3,4,5])
print(*a)
```

```python
# extend array and insert multiple element at once

a.extend([6,7,8,9,10])
print(*a)
```
